```python
def ySharpe_Ratio(df, columns, risk_free_rate=0):
    
    df.index = pd.to_datetime(df.index)
    yearly_df = df.resample('Y').sum()
    exp_ret = yearly_df[columns].mean()
    std = yearly_df[columns].std()
    
    Sharpe_ratio = (exp_ret - risk_free_rate) / std
    
    return Sharpe_ratio, exp_ret, std
```

# Montecarlo Optimization

```python
sec_df = merged_df.iloc[:,:4]


# Number of securities
num_of_sec = 4

# Number of times we want to run the simulation
num_of_port = 3000

# Prepare the weight arrays of np.zeros
weights_arr = np.zeros((num_of_port, num_of_sec))

# Prepare the return arrays of numpy zeros
ret_arr = np.zeros(num_of_port)

# Prepare the std arrays of numpy zeros 
std_arr = np.zeros(num_of_port)

# Prepare the Sharpe rario arrays of numpy zeros 
sharpe_arr = np.zeros(num_of_port)

# Run the simulation
for ind in range(num_of_port):
    
    # First, calculate the weights and normalize them
    weights = np.array(np.random.random(num_of_sec))
    weights = weights / np.sum(weights)
    
    # Then, add weights rwo to weights array
    weights_arr[ind, :] = weights
    
    # Calculate the expected annual returns
    ret_arr[ind] = sec_df.dot(weights).mean() * 252
    
    # Calculate the expected annual std (w.T * varcov matrix * w)
    std_arr[ind] = np.sqrt(np.dot(weights.T, np.dot((sec_df.cov() * 252), weights)))
   
    # Calculate the Sharpe ratio
    sharpe_arr[ind] = (ret_arr[ind]) / std_arr[ind] 
        
def format_weights(weights_arr):
    return [f'{weight:.3f}' for weight in weights_arr]

formatted_weights_arr = [format_weights(weights) for weights in weights_arr]


# Combine them all together
simulations_data = [ret_arr, std_arr, sharpe_arr, formatted_weights_arr]
simulations_df = pd.DataFrame(data=simulations_data).T
simulations_df.columns = ['Returns', 'Standard Deviation', 'Sharpe Ratio', 'Weights']
simulations_df = simulations_df.infer_objects()


    
    
```

```python
def get_metrics(weights:list) -> np.array:
    
    # Convert weights to numpy array
    weights = np.array(weights)
    
    # Calculate returns
    ret = sec_df.dot(weights).mean() * 252
    
    # Calculate std
    std = np.sqrt(np.dot(weights.T, np.dot((sec_df.cov() * 252), weights)))
    
    # Calculate Sharoe Ratio
    sharpe_ratio = ret / std
    
    return np.array([ret, std, sharpe_ratio])


def negative_sharp(weights:list) -> np.array:
    
    return get_metrics(weights[2]) * -1

def grab_volatility(weights:list) -> np.array:
    
    return get_metrics(weights[1])

def check_sum(weights:list) -> float:
    
    return np.sum(weights) - 1
    
```
